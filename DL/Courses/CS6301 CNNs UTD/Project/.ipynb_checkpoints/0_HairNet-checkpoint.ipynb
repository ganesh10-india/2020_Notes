## Visualizations

```python
def sigmoid(x):
    return tf.divide(1,tf.add(tf.exp(-x),1))
func_name='Sigmoid'
func=sigmoid

x = tf.range(-3.,3, delta=.25)
with tf.GradientTape() as t:
    t.watch(x)
    y = func(x)
    

gradient = t.gradient(y, x) 

plt.title("{} Activation".format(func_name))
plt.plot( x.numpy(), y.numpy(), c='black')
plt.ylabel("{}(x)".format(func_name))
plt.xlabel("X")
plt.axvline(x=-1,c='grey', alpha=0.7)
plt.axvline(x=1,c='grey', alpha=0.7)
plt.show()

plt.title("{} Gradient".format(func_name))
plt.plot( x.numpy(), gradient.numpy(), c='black')
plt.ylabel("delta_{}/delta_x".format(func_name))
plt.xlabel("X")
plt.axvline(x=-1,c='grey', alpha=0.7)
plt.axvline(x=1,c='grey', alpha=0.7)
plt.show()
```

```python
def dwise_block_params(in_dims, out_dims, repeats):
    params = 3*3*in_dims+ 1*1*out_dims*in_dims
    mid_params = 3*3*out_dims+ 1*1*out_dims*out_dims
    return params+(repeats-1)*mid_params
dwise_block_params(512, 1024, 2)
```

## CNN Architectures
### AlexNet

```python
# model
MODEL_LEVEL_0_BLOCKS    = 4
MODEL_LEVEL_1_BLOCKS    = 4
MODEL_LEVEL_2_BLOCKS    = 5
BLOCK_REPEATS = [MODEL_LEVEL_0_BLOCKS, MODEL_LEVEL_1_BLOCKS, MODEL_LEVEL_2_BLOCKS]


def hairnet_tail(inputs, dims):
    out = conv_block(inputs, dims, kernel_size=(3,3), strides=(1,1), activation=True)
    return out

def depthwise_sep_conv(inputs, dims, **kwargs):
    """
    inputs: Tensor- input to the first layer
    dims: int - depth of the channel dimension 
    
    Depthwise seperable convolution a la MobileNet V1.
    """
    x = DepthwiseConv2D((3,3), strides=(1,1), **conv_params)(inputs) 
    x = BatchNormalization(**bn_params)(x)
    x = ReLU()(x)
    
    x = Conv2D(dims, (1, 1), strides=(1, 1), **conv_params)(x) 
    x = BatchNormalization(**bn_params)(x) 
    x = ReLU()(x)
    return x

def hairnet_head(inputs, dims=None):
    x = conv_block(inputs, DATA_NUM_CLASSES, kernel_size=(3,3), strides=(1,1), activation=True)
    x = GlobalAveragePooling2D()(x)
    x = tf.keras.activations.softmax(x)
    return x
```

```python
hairnet = VGG_Like_CNN(hairnet_tail,
            depthwise_sep_conv,
            hairnet_head,
            input_shape = (DATA_CROP_ROWS, DATA_CROP_COLS, DATA_CHANNELS),
            num_levels= 3,
            num_downsamples=2,
            block_repeats=BLOCK_REPEATS)

hairnet.compile(optimizer = tf.keras.optimizers.SGD(TRAINING_LR_MAX),
              loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])
```

```python
TRAINING_LR_MAX          = 0.006
hist = train(hairnet, dataset_train, dataset_test, 'hairnet', logs=False, save=True)
```

### HairNet without BN

```python
def hairnet_tail_no_BN(inputs, dims):
    x = Conv2D(dims, (3,3), activation='relu', padding='same')(inputs)
    return x

def depthwise_sep_conv_no_BN(inputs, dims, **kwargs):
    """
    inputs: Tensor- input to the first layer
    dims: int - depth of the channel dimension 
    
    Depthwise seperable convolution a la MobileNet V1.
    """
    x = DepthwiseConv2D((3,3), strides=(1,1), **conv_params)(inputs) 
    x = ReLU()(x)
    
    x = Conv2D(dims, (1, 1), strides=(1, 1), **conv_params)(x) 
    x = ReLU()(x)
    return x

def hairnet_head_no_BN(inputs, dims=None):
    x = Conv2D(DATA_NUM_CLASSES, (3,3), activation='relu', padding='same')(inputs)
    x = GlobalAveragePooling2D()(x)
    x = tf.keras.activations.softmax(x)
    return x
```

```python
hairnet = VGG_Like_CNN(hairnet_tail_no_BN,
            depthwise_sep_conv_no_BN,
            hairnet_head_no_BN,
            input_shape = (DATA_CROP_ROWS, DATA_CROP_COLS, DATA_CHANNELS),
            num_levels= 3,
            num_downsamples=2,
            block_repeats=BLOCK_REPEATS)

hairnet.compile(optimizer = tf.keras.optimizers.SGD(TRAINING_LR_MAX),
              loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])

hist = train(hairnet, dataset_train, dataset_test, 'hairnet_no_BN', logs=False, save=True)
```

### With a final Dense Layer

```python
def hairnet_dense_head(inputs, dims=None):
    x = GlobalAveragePooling2D()(inputs)
    x = Dense(DATA_NUM_CLASSES, activation='softmax')(x)
    return x
```

```python
hairnet = VGG_Like_CNN(hairnet_tail,
            depthwise_sep_conv,
            hairnet_dense_head,
            input_shape = (DATA_CROP_ROWS, DATA_CROP_COLS, DATA_CHANNELS),
            num_levels= 3,
            num_downsamples=2,
            block_repeats=BLOCK_REPEATS)

hairnet.compile(optimizer = tf.keras.optimizers.SGD(TRAINING_LR_MAX),
              loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])

hist = train(hairnet, dataset_train, dataset_test, 'hairnet_w_dense', logs=False, save=True)
```

### With Residual Connections

```python
def depthwise_sep_conv_residual(inputs, dims, downsample=False, **kwargs):
    """
    inputs: Tensor- input to the first layer
    dims: int - depth of the channel dimension 
    
    Depthwise seperable convolution a la MobileNet V1.
    """   
    x = inputs
    
    resid = DepthwiseConv2D((3,3), strides=(1,1), **conv_params)(inputs) 
    resid = BatchNormalization(**bn_params)(resid)
    resid = ReLU()(resid)
    
    resid = Conv2D(dims, (1, 1), strides=(1, 1), **conv_params)(resid) 
    resid = BatchNormalization(**bn_params)(resid) 
    
    if downsample:
        x = Conv2D(dims, (1,1))(x)
        
    out = Add()([resid,x])
    out = ReLU()(out)
    return out
```

```python
hairnet = VGG_Like_CNN(hairnet_tail,
            depthwise_sep_conv_residual,
            hairnet_head,
            input_shape = (DATA_CROP_ROWS, DATA_CROP_COLS, DATA_CHANNELS),
            num_levels= 3,
            num_downsamples=2,
            block_repeats=BLOCK_REPEATS)

hairnet.compile(optimizer = tf.keras.optimizers.SGD(TRAINING_LR_MAX),
              loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])

hist = train(hairnet, dataset_train, dataset_test, 'hairnet_w_residual', logs=False, save=True)
```

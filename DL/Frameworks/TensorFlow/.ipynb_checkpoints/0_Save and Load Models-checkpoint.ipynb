### Save and Load Models


```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

train_labels = train_labels[:1000]
test_labels = test_labels[:1000]

train_images = train_images[:1000].reshape(-1, 28 * 28) / 255.0
test_images = test_images[:1000].reshape(-1, 28 * 28) / 255.0

# Define a simple sequential model
def create_model():
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(512, activation='relu', input_shape=(784,)),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(10, activation='softmax')
        ])
    model.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])
    return model

# Create a basic model instance
model = create_model()
```

The `tf.keras.callbacks.ModelCheckpoint` callback allows to continually save the model both during and at the end of training.

```python
checkpoint_path = "C://Users/harri/.keras/models/model_save_test_MNIST_2-3-20/cp.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)
```

```python
checkpoint_dir
```

```python
# Create a callback that saves the model's weights
cp_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_dir,
                                                 save_best_only=True,
                                                 verbose=1)
# Train the model with the new callback
model.fit(train_images, 
          train_labels,  
          epochs=10,
          validation_data=(test_images,test_labels),
          callbacks=[cp_callback])  # Pass callback to training
```

```python
# Create a basic model instance
model = create_model()

# Evaluate the model
loss, acc = model.evaluate(test_images,  test_labels, verbose=2)
print("Untrained model, accuracy: {:5.2f}%".format(100*acc))
```

```python
# Loads the weights
model = tf.keras.models.load_model(checkpoint_dir)

# Re-evaluate the model
loss, acc = model.evaluate(test_images,  test_labels, verbose=2)
print("Restored model, accuracy: {:5.2f}%".format(100*acc))
```

Call `model.save` to save the a model's architecture, weights, and training configuration in a single file/folder. This allows you to export a model so it can be used without access to the original Python code*. Since the optimizer-state is recovered, you can resume training from exactly where you left off.

```python
# Save the entire model to a HDF5 file.
# The '.h5' extension indicates that the model shuold be saved to HDF5.
model.save('my_model.h5') 
```

```python
model=tf.keras.models.load_model('my_model.h5')
# Re-evaluate the model
loss, acc = model.evaluate(test_images,  test_labels, verbose=2)
print("Restored model, accuracy: {:5.2f}%".format(100*acc))
```

### Load Image Data


```python
import IPython.display as display
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import pathlib
```

`tf.keras.preprocessing.image.ImageDataGenerator`-  generates batches of Tensor image data, augmented in real time. `.ImageDataGenerator.flow_from_directory()` will generate batches from a dirctory of images. 

`tf.keras.utils.get_file` downloads a file from the internet

`pathlib.Path` objects can be extended with the operator / as well as `.glob(regex)` to find files in subdirectories

```python
img_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)

data_dir = tf.keras.utils.get_file(origin='https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz',
                                         fname='flower_photos', untar=True)
data_dir = pathlib.Path(data_dir)
image_count = len(list(data_dir.glob('*/*.jpg')))

BATCH_SIZE = 32
IMG_HEIGHT = 224
IMG_WIDTH = 224
STEPS_PER_EPOCH = np.ceil(image_count/BATCH_SIZE)
CLASS_NAMES = np.array([item.name for item in data_dir.glob('*') if item.name != "LICENSE.txt"])

train_data_gen = img_generator.flow_from_directory(directory=str(data_dir),
                                                   batch_size = BATCH_SIZE,
                                                   shuffle=True,
                                                   target_size = (IMG_HEIGHT, IMG_WIDTH),
                                                   classes = list(CLASS_NAMES))
```

```python
data_dir
```

```python
def show_batch(image_batch, label_batch):
    plt.figure(figsize=(10,10))
    for n in range(25):
        ax = plt.subplot(5,5,n+1)
        plt.imshow(image_batch[n])
        plt.title(CLASS_NAMES[label_batch[n]==1][0].title())
        plt.axis('off')
image_batch, label_batch = next(train_data_gen)
show_batch(image_batch, label_batch)
```

#### Load Images using tf.data
The above `keras.preprocessing` method is convienient, but has three downsides:

- It's slow. See the performance section below.
- It lacks fine-grained control.
- It is not well integrated with the rest of TensorFlow.

To load the files as a `tf.data.Dataset` first create a dataset of the file paths:

```python
!ls {data_dir}
```

`tf.data.Dataset.list_files()` - The default behavior of this method is to return filenames in a non-deterministic random shuffled order

```python
list_ds = tf.data.Dataset.list_files(str(data_dir/'*/*'))
```

This is a list of filepaths, we need a function that converts a filepath to (image_data, label) pair

```python
def get_label(file_path):
    parts = tf.strings.split(file_path, os.path.sep)
    return parts[-2] == CLASS_NAMES

def decode_image(img):
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    return tf.image.resize(img, [IMG_WIDTH, IMG_HEIGHT])

def process_path(file_path):
    label = get_label(file_path)
    img = tf.io.read_file(file_path)
    img = decode_image(img)
    return (img, label)

labeled_ds = list_ds.map(process_path) #maps the function to all filepaths in the dataset
```

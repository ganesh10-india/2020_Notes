## For Images
Image dataset is Tiny ImageNet [source](https://tiny-imagenet.herokuapp.com/)

```python
dir_path = "E://Data/Imagenet/tiny-imagenet-200/"
NUM_CLASSES = 200
BATCH_SIZE=128
EPOCHS=5

def show_batch(train_data_gen):
    """Implementing something similar to FastAIs showbatch method
    Outputs a 3x3 grid of images from the image data generator"""
    image_batch, label_batch = next(train_data_gen)
    plt.figure(figsize=(10,10))
    CLASSES = np.array(list(train_data_gen.class_indices.keys()))  #get the label order directly from iterator (otherwise will mixup)
    for n in range(9):
        ax = plt.subplot(3,3,n+1)
        plt.imshow(image_batch[n])
        plt.title(CLASSES[label_batch[n]==1][0].title())
        plt.axis('off')
    plt.show()
    
    
def timeit(ds, steps=100):
    """
    Timing method for data pipeline iterator.
    Outputs the time for one-thousand calls to the iterator
    """
    start = time.time()
    it = iter(ds)
    for i in range(steps):
        batch = next(it)
        if i%10 == 0:
            print('.',end='')
    print()
    end = time.time()
    duration = end-start
    print("{} batches: {} s".format(steps, duration))
    print("{:0.5f} Images/s".format(BATCH_SIZE*steps/duration))

    
def Example_CNN_Model(num_blocks, convs_per_block):
    """
    Example Convolutional Neural Network for Image data.
    Example_CNN_Model(num_blocks, convs_per_block)
    """
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Input(shape=(64,64,3),
                                   batch_size=BATCH_SIZE))
    filters = 64
    for i in range(num_blocks):
        for j in range(convs_per_block):
            model.add(tf.keras.layers.Conv2D(filters*(i+1),
                                                  (3,3), 
                                                  padding='same',
                                                  activation='relu'))
            model.add(tf.keras.layers.BatchNormalization())
        model.add(tf.keras.layers.MaxPooling2D())
    model.add(tf.keras.layers.GlobalAvgPool2D())
    model.add(tf.keras.layers.Dense(NUM_CLASSES, activation="softmax"))
    model.compile(optimizer='adam',
                 loss='categorical_crossentropy',
                 metrics=['acc'])
    return model
```

```python
data_root = os.path.join(dir_path,"train")
data_root = pathlib.Path(data_root)
CLASSES = np.array(os.listdir(data_root))

print("Example classes:",CLASSES[:5])
print("Number of classes:",len(CLASSES))
```

### Keras `ImageDataGenerator` API

Easy to use, directory-based image processing API. Allows a host of image augmentations to be applied to input images. Slowere than the tf.data API.

```python
#define the tranformations
transforms = {
            'validation_split':0.1,
              "rescale":1./255
             }

#Create the generator with spectified transforms
image_gen = tf.keras.preprocessing.image.ImageDataGenerator(**transforms)

#create generators from data directory
keras_train_imgs =  image_gen.flow_from_directory(data_root,
                                           target_size = (64,64),
                                           batch_size = BATCH_SIZE,
                                            subset='training')

keras_valid_imgs =  image_gen.flow_from_directory(data_root,
                                           target_size = (64,64),
                                           batch_size = BATCH_SIZE,
                                            subset='validation')
```

```python
show_batch(keras_train_imgs)
```

#### Model Training

```python
model = Example_CNN_Model(3, 2)
model.summary()
```

```python
STEPS_PER_EPOCH = keras_train_imgs.samples//BATCH_SIZE
#VALID_STEPS_PER_EPOCH = valid_imgs.samples//BATCH_SIZE
start_time = time.time()

#Model checkpointing during training.
checkpoint_path = 'E://Models/ImageNet-200/1/'
cb = [tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                                 save_best_only=True,
                                                 verbose=1)]

#use_multiprocessing on generator makes training faster
history = model.fit(keras_train_imgs,
           steps_per_epoch=STEPS_PER_EPOCH,
           epochs=EPOCHS,
           validation_data=keras_valid_imgs,
           callbacks=cb,
           workers=-1,
           use_multiprocessing=True)
                                         
print("--- %s seconds ---" % (time.time() - start_time))
```

### `tf.data` API 

One of the biggest time savers of the `tf.data.Dataset` API is the option to `prefetch` which allows data to be loaded asynchronously, and the option to `cache` which allows data to be retained in memory for faster access on subsequent references.

Most of the speed gains come from caching. If the dataset does not fit in memory, retaining a cache file will allow some benefit (filename ebds with `.tfcache`). Pass the name a filepath to the `.cache()` method.

#### Sources
- [load images doc](https://www.tensorflow.org/tutorials/load_data/images#load_using_tfdata)

```python
#Need a function to map to the datafle dataset above
def get_label(file_path):
  # convert the path to a list of path components
  parts = tf.strings.split(file_path, os.path.sep)
  # The second to last is the class-directory
  return parts[-2] == CLASSES

def decode_image(img):
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    return img

def process_path(path):
    """
    Input: file_path of a sample image
    Output: image in 3x64x64 float32 Tensor and one hot tensor
    """
    label = get_label(path)
    image = tf.io.read_file(path)
    image = decode_image(image)
    return image, label


def prepare_for_training(ds, cache=True, shuffle_buffer_size=1000):
    # This is a small dataset, only load it once, and keep it in memory.
    # use `.cache(filename)` to cache preprocessing work for datasets that don't
    # fit in memory.
    if cache:
        if isinstance(cache, str):
            ds = ds.cache(cache)
        else:
            ds = ds.cache()
    ds = ds.shuffle(buffer_size=shuffle_buffer_size)
    # Repeat forever
    ds = ds.repeat()
    ds = ds.batch(BATCH_SIZE)
    # `prefetch` lets the dataset fetch batches in the background while the model
    # is training.
    ds = ds.prefetch(buffer_size=AUTOTUNE)
    return ds
```

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE

list_files = tf.data.Dataset.list_files(str(data_root/'*/*'))

#map the above function to file_name dataset
train_imgs = list_files.map(process_path, num_parallel_calls=AUTOTUNE) 
train_ds_nocache = prepare_for_training(train_imgs, cache=False, shuffle_buffer_size=1000)
train_ds_cachefile = prepare_for_training(train_imgs, cache="E://Models/ImageNet-200/cache.tfcache", shuffle_buffer_size=1000)
train_ds_cache = prepare_for_training(train_imgs, cache=True, shuffle_buffer_size=1000)
```

```python
#working with a directory of txt files
DIRECTORY_URL = 'https://storage.googleapis.com/download.tensorflow.org/data/illiad/'
FILE_NAMES = ['cowper.txt', 'derby.txt', 'butler.txt']

for name in FILE_NAMES:
    text_dir = tf.keras.utils.get_file(name, origin=DIRECTORY_URL+name)

parent_dir = os.path.dirname(text_dir)
print(parent_dir)
```

```python
def labeler(example, index):
    return example, tf.cast(index, tf.int64) 

labeled_data_sets = []

for i, file_name in enumerate(FILE_NAMES):
    lines_dataset = tf.data.TextLineDataset(os.path.join(parent_dir, file_name))
    labeled_dataset = lines_dataset.map(lambda ex: labeler(ex, i))
    labeled_data_sets.append(labeled_dataset)
```

```python
BUFFER_SIZE = 50000
BATCH_SIZE = 64
TAKE_SIZE = 5000

all_labeled_data = labeled_data_sets[0]
for labeled_dataset in labeled_data_sets[1:]:
    all_labeled_data = all_labeled_data.concatenate(labeled_dataset)
    
all_labeled_data = all_labeled_data.shuffle(BUFFER_SIZE, reshuffle_each_iteration=False)
```

```python
encoder = tfds.features.text.TokenTextEncoder(vocabulary_set)

def encode(text_tensor, label):
    encoded_text = encoder.encode(text_tensor.numpy())
    return encoded_text, label

def encode_map_fn(text, label):
    # py_func doesn't set the shape of the returned tensors.
    encoded_text, label = tf.py_function(encode, 
                                       inp=[text, label], 
                                       Tout=(tf.int64, tf.int64)) 
    encoded_text.set_shape([None])
    label.set_shape([])
    return encoded_text, label


all_encoded_data = all_labeled_data.map(encode_map_fn)
```

## Other Useful Functions

- `tf.data.Dataset.from_generator()` constructor generates a Dataset object from a Python generator.
    - Can use this to wrap a `tf.keras.preprocessing.image.ImageDataGenerator`
    - Make sure to specifythe ouput types and output shapes
- `tf.keras.utils.get_file()` - downloads a file from the internet with the option to untar
- `tf.data.Dataset.list_files(str(flowers_root/'*/*'))` - Use path regex to collect all the files within a directory.
- `tf.io.read_file()` - reads the entire contents of a given file path

- `Dataset.batch()` - results in a Tensor with unknown batch size because the last batch may not be full. 
    - use `drop_remainder` to get full shape propagation. 
- `Dataset.padded_batch()` allows batching of tensors with different lengths. By specifying the dimensions to be padded

### Best Practice Summary ([link](https://www.tensorflow.org/guide/data_performance#best_practice_summary))
- Use the `.prefetch()` transformation to overlap data collection and consumption.
- Parallelize data reading transformation using `.interleave()`
- Parallelize the `map` transformation with `num_parallel_calls`
- `cache` data after the `map` function
- Vectorize the `map` function and apply after the `batch`ing

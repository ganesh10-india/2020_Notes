### Sampling Process
1. Define the Population of concern
2. Specify a sampling frame - set of items possible to measure
3. Determine the sample size
4. Implement the sampling plan
5. Sampling and collecting data
6. Review the sampling procsess



### Simple Random Sampling

Applicable __when population is small, homogeneous & readily available.__ All subsets of sampling frame are given an equal prob. to be sampled.

__Disadvantages__ Impractical if the sampling frame is large. Minority subgroups may be undersampled.

### Systematic Sampling
 Arrange the target population according to some ordering scheme, then select elements at regular intervals. Involves a randomized start and then selects the kth element from then onwards. __sampling spread evenly across ref. population.__
 
 __Disadvantages__ Biased if there is a hidden periodicity that coincides with assignment.

### Stratified Sampling
When population has a number of distinct categories, seperate the frame into strata. Each strata is sampled as an independent subpopulation. Use sampling fraction for strata to ensure proportianate representation in the sample. DIfferent sampling approaches can be applied to different strata.

__Disadvantages__ May require a larger sample than other methods (if min sample size per group)

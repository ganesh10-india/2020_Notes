```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("counts.csv")
df.drop("Unnamed: 0", axis=1,inplace=True)
df
```

```python
df["n"]=0
```

```python
ns = [5,10,30,100,150,200]
for i in range(6):
    df.iloc[i*4:4+i*4,3]=ns[i]
```

```python
df
```

```python
lambdas = [0.01,0.1,1,10]
for i in range(4):
    _lambda = lambdas[i]
    subdf = df[df['lambda']==_lambda]
    plt.figure(figsize=(6,4))
    plt.title("Coverage Probability for Lambda = "+str(_lambda))
    plt.ylabel("Percent Covered")
    plt.xlabel("Obsvs in Sample")
    plt.ylim(0.75,1)
    plt.xlim(5,200)
    plt.plot(subdf.n,subdf.z_est,label="Z-Interval")
    plt.plot(subdf.n,subdf.b_est,label="Boot-Interval")
    plt.hlines(0.95,5,200,colors="black",linestyle="dashed",alpha=0.3)
    plt.text(20,0.955, "95% Coverage", alpha=0.3)
    plt.legend(loc=4)
```

```python
for i in range(2):
    plt.figure(figsize=(6,4))
    if i==0:
        plt.title("Coverage Probability for Bootstrapped-Interval")
    else:
        plt.title("Coverage Probability for Z-Interval")
    plt.ylabel("Percent Covered")
    plt.xlabel("Obsvs in Sample")
    plt.ylim(0.75,1)
    plt.xlim(5,200)
    plt.hlines(0.95,5,200,colors="black",linestyle="dashed",alpha=0.3)
    plt.text(20,0.955, "95% Coverage", alpha=0.3)

    for j in range(4):
        _lambda = lambdas[j]
        subdf = df[df['lambda']==_lambda]
        if i==0:
            plt.plot(subdf['n'],subdf.b_est,label="lambda="+str(_lambda))
        else:
            plt.plot(subdf['n'],subdf.z_est,label="lambda="+str(_lambda))
        plt.legend(loc=4)
        
```
